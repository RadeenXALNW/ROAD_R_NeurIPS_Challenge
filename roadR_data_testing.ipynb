```python
import os

def count_images_in_folder(folder_path, image_extensions=[".jpg", ".jpeg", ".png", ".gif"]):
    total_images = 0

    for root, dirs, files in os.walk(folder_path):
        for file in files:
            if any(file.lower().endswith(ext) for ext in image_extensions):
                total_images += 1

    return total_images

main_folder_path = '/content/road/rgb-images'
total_images = count_images_in_folder(main_folder_path)

print(f"Total number of images under '{main_folder_path}': {total_images}")
```

```python
import os

def get_folder_size(folder_path):
    total_size = 0

    for root, dirs, files in os.walk(folder_path):
        for file in files:
            file_path = os.path.join(root, file)
            total_size += os.path.getsize(file_path)

    return total_size

def format_size(size_bytes):
    # Function to format size in a human-readable format
    power = 2**10
    n = 0
    size_labels = {0: '', 1: 'KB', 2: 'MB', 3: 'GB', 4: 'TB'}

    while size_bytes > power:
        size_bytes /= power
        n += 1

    return f"{size_bytes:.2f} {size_labels[n]}"

main_folder_path = '/content/road/rgb-images'
total_size_bytes = get_folder_size(main_folder_path)
total_size_formatted = format_size(total_size_bytes)

print(f"Total size of images under '{main_folder_path}': {total_size_formatted}")
```
